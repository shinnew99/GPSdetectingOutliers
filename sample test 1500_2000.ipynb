Accuracy = GPS가 보내는 정확도

```python
import pandas as pd
import folium

import skmob
from skmob.preprocessing import filtering
```

```python
def osm_visualization(df):
    df = df.reset_index(drop=True)
    map_osm = folium.Map(location=[df.latitude.mean(), df.longitude.mean()], zoom_start=13)

    for i in df.index:
        marker = (df.latitude[i], df.longitude[i])
        folium.CircleMarker(marker, radius=5,
                            fill=True, fill_color='red', fill_opacity=0.75,
                            weight=1, color='black').add_to(map_osm)
    return map_osm
```

```python
df = pd.read_csv('move.csv')
df['uid'] = 1
tdf = skmob.TrajDataFrame(df, latitude='latitude', longitude='longitude', user_id='uid', datetime='datetime')
```
