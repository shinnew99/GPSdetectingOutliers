```python
from skmob.preprocessing import filtering
```

- filtering 시 loops를 include하고, max= 8로 했을 때 일단 outlier들은 다듬어 지는데..

```python
ftdf290 = filtering.filter(tdf, max_speed_kmh=290, include_loops=True, max_loop=6, ratio_max=0.5)
ftdf290.plot_trajectory()
```

```python
from skmob.preprocessing import detection
```

```python
stdf = detection.stops(ftdf290, stop_radius_factor=0.5, minutes_for_a_stop=20.0, spatial_radius_km=0.2, leaving_time=True)
stdf.plot_stops()
```
